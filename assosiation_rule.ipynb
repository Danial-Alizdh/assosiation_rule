```python
def create_SCL(df, k):
    columns = df.columns.values.tolist()
    columns.remove('Class')
    normalize = False
    subsets = list(itertools.combinations(columns, k)).copy()
    columns = []
    class_columns = []
    support = []
    class_support = []
    for subset in subsets:
        columns_list = list(subset)
        lenght = len(columns_list)
        value_counts = df[columns_list].value_counts(normalize=normalize)
        if len(value_counts) % (2**lenght) == 0:
            for i in range(lenght):
                value_counts = value_counts[1]
            support.append(value_counts)
        else:
            support.append(0)
        v = ""
        l = len(columns_list) - 1
        for i in range(l):
            v += columns_list[i] + ', '
        columns.append(v + columns_list[l])

    support.append(df['Class'].value_counts(normalize=normalize)[1])
    columns.append('Class')
    df_support = pd.DataFrame(columns=columns)
    df_support.loc['Support'] = support
    threshold1 = 400
    threshold2 = 14200
    thre_list = df_support.loc['Support'] > threshold2
    thre_list2 = df_support.loc['Support'] < threshold1
    df_support.drop(columns=[df_support.columns[i] for i in [i for i in range(len(thre_list)) if thre_list[i] == True]], inplace=True)
    df_support.drop(columns=[df_support.columns[i] for i in [i for i in range(len(thre_list2)) if thre_list2[i] == True]], inplace=True)

    columns = df_support.columns
    for col in columns:
        if col != 'Class':
            cols = list(col.split(', '))
            columns_list = cols.copy()
            if 'Class' not in cols:
                columns_list.append('Class')
            lenght = len(columns_list)
            value_counts = df[columns_list].value_counts(normalize=normalize)
            if len(value_counts) % (2**lenght) == 0:
                for i in range(lenght):
                    value_counts = value_counts[1]
                class_support.append(value_counts)
            else:
                class_support.append(0)
            v = ""
            l = len(columns_list) - 1
            for i in range(l):
                v += columns_list[i] + ', '
            class_columns.append(v + columns_list[l])
    
    df_class_support = pd.DataFrame (columns=class_columns)
    df_class_support.loc['Class_support'] = class_support

    columns = df_class_support.columns
    confidence = []
    lift = []
    indexs = []
    for col in columns:
        if col != 'Class':
            cols = list(col.split(', '))
            if 'Class' in cols:
                cols.remove('Class')
            c = ""
            for i in range(len(cols) - 1):
                if cols[i] != 'Class':
                    c += cols[i] + ', '
            if len(cols) > 0:
                c += cols[len(cols) - 1]
            indexs.append(c)
            if c in df_support.columns:
                supX = df_support[c][0]
            else: 
                supX = 0
            supY = df_support['Class'][0]
            supXY = df_class_support[col][0]
            if supX == 0 or supXY == 0:
              confidence.append(0)
              lift.append(0)
            else:
              confidence.append(supXY / supX)
              lift.append((supXY) / (supX * supY))
              
    if len(indexs) > 0:
        df_confidence_lift = pd.DataFrame (columns=[v + ' -> Class' for v in indexs])
        df_confidence_lift.loc['Confidence'] = confidence
        df_confidence_lift.loc['Lift'] = lift
    else:
        df_confidence_lift = pd.DataFrame()

    return df_support, df_class_support, df_confidence_lift
```
